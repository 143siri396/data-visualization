```python
#line plot
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
x=np.arange(0,10,1)
y=2*x+5
plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
x=np.arange(0,10,1)
y=2*x+5
plt.plot(x,y,color="lime",marker="*")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
x=np.arange(0,10,1)
y=2*x+5
fig=plt.figure(figsize=(10,5))
plt.plot(x,y,linewidth=2.0,color="black",marker="*",linestyle="-.",alpha=0.9,markerfacecolor='r')
plt.xlabel("x-asis")
plt.ylabel("y-axis")
plt.title("line plot")
plt.legend(['line1'],loc='best')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
x=np.arange(0,10,1)
y1=2*x+5
y2=3*x+10
plt.subplot(1,2,1)
plt.plot(x,y1,color="red",marker="*")
plt.title('graph1')
plt.subplot(1,2,2)
plt.plot(x,y2,color="y",marker="*")
plt.title('graph2')
plt.show()
```

```python
#bar plot
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
data={"Apple":20,"Banana":10,"Mango":15,"lemon":30}
names=list(data.keys())
values=list(data.values())
plt.bar(names,values,color=["r","b","c","black"])
plt.title("bar chart")
plt.show()
```

```python
#bar plot
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
data={"Apple":20,"Banana":10,"Mango":15,"lemon":30}
names=list(data.keys())
values=list(data.values())
plt.barh(names,values,color=["r","b","c","black"])
plt.title("bar chart")
plt.show()
```

```python
#scatterplot
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
a=[10,20,30,40,50,60,70,80]
b=[5,3,2,5,6,1,4,2]
x=[1,2,3,4,3,2,1,2]
plt.scatter(a,x,color="green",marker="*",alpha=0.9)
plt.scatter(b,x,color="red",marker="*",alpha=0.9)
plt.title("scatterplot")
plt.ylabel('y-axis')
plt.xlabel('x-axis')
plt.grid(True)
plt.savefig("scatterplot.png")
plt.show()

```

```python
#histogram
import matplotlib.pyplot as plt
age=[10,12,16,19,11,20,26,28,38,35,34,45,60,55,56,58,68,78,23,89,87,90,99,96,98]
plt.hist(age,bin=[0,20,40,60,80,100])
plt.show()
```

```python
#piechart
import matplotlib.pyplot as plt
animal=['dog','cat','wolf','donkey','lion']
count=[40,50,60,20,70]
plt.pie(count,labels=animal,autopct="%1.1f%%",shadow=True,startangle=90,explode=(0,0,0.1,1,0))
plt.title("piechart")
plt.show()
```

```python
#donut chart
import matplotlib.pyplot as plt
animal=['dog','cat','wolf']
count=[40,20,70]
size_centre=[10]
pie1=plt.pie(count,labels=animal,autopct="%1.1f%%",shadow=True,startangle=90)
pie2=plt.pie(size_centre,radius=1.0,colors="white")
plt.show()
```

```python
#area chart
import matplotlib.pyplot as plt
import numpy as np
x=range(1,15)
y=[1,4,6,8,4,5,3,6,8,9,5,4,5,6]
plt.stackplot(x,y,colors="green",alpha=1.0)
plt.plot(x,y,color="k",marker="*")
plt.grid(True)
plt.show()
```
